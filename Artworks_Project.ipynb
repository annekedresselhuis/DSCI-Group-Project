```r
library(tidyverse)
```

```r
Artworks <- read_csv("https://media.githubusercontent.com/media/MuseumofModernArt/collection/master/Artworks.csv")
```

```r
glimpse(Artworks)
```

```r
artworks <- Artworks 
colnames(artworks) <- make.names(colnames(artworks), unique = TRUE)
artworks
        
```

```r
artworks_selected <- artworks %>%
        select(-ThumbnailURL, -URL, -AccessionNumber, -ConstituentID, -Circumference..cm., -Cataloged, -ObjectID, -Depth..cm., -Diameter..cm., -Length..cm., -Weight..kg., -Seat.Height..cm., -Duration..sec..)
artworks_selected
```

```r
artworks_selected$Area<- with(artworks_selected, Height..cm.*Width..cm.)
artworks_selected
```

```r
artworks_size_relation <- artworks_selected %>%
    select(Gender, Area, Date)
artworks_size_relation
```

```r
artworks_male <- artworks_size_relation %>%
    filter(Gender=="(Male)")
artworks_male

artworks_female <- artworks_size_relation %>%
    filter(Gender=="(Female)")
artworks_female

nrow(artworks_male)
nrow(artworks_female)

#about a 1:6 ratio of female to male
#some of the dates (years) recorded in different format, how do we change them or filter them out?
```

```r
male_art_plot <- artworks_male%>%
        ggplot(aes(x=Date, y=Area))+
        geom_point()+
        labs(x="Year", y="Artwork Area (cm^2)")
male_art_plot

female_art_plot <- artworks_female%>%
        ggplot(aes(x=Date, y=Area))+
        geom_point()+
        labs(x="Year", y="Artwork Area (cm^2)")
female_art_plot

#How do I rescale axis??? This is illegible. 
```
