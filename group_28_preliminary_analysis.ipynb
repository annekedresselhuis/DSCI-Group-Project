```r
library(tidyverse)
library(lubridate)

#load dataset
artworks <- read_csv("https://media.githubusercontent.com/media/MuseumofModernArt/collection/master/Artworks.csv")

#explore a subset of the data: last 100 entries
artworks_100 <- tail(artworks, 100)
artworks_100
```

```r
#tidy column names
colnames(artworks) <- make.names(colnames(artworks), unique = TRUE)
```

```r
# create a new column for artwork surface area
artworks <- mutate(artworks, surface_area = Width..cm. * Height..cm.)

# transform column types
artworks <- mutate(artworks, Date = as.Date(Date), na.rm = TRUE) # Date as a date variable
artworks <- mutate(artworks, EndDate = as.Date(EndDate, format = "(%Y)", na.rm = TRUE)) # EndDate as a date variable
artworks <- mutate(artworks, Gender = as_factor(Gender) # Gender as a factor variable


# create a plot of surface area over time

artworks_plot <- artworks %>%
ggplot(aes(x = Date, y = surface_area)) +
geom_point() +
labs(x = "Year", y = "Surface Area (cm^2)")

artworks_plot 
                   
#got stuck here ?
```
